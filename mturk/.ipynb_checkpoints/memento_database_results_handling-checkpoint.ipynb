```python
def show_distribution(arr, title='Data', distplot=False, bins = 20, vert_bar_at=None, figsize=[10,7]):
    plt.figure(figsize=figsize)
    if distplot:
        sns.distplot(arr)
    else:
        plt.hist(arr, bins = bins)
        
    if vert_bar_at:
        plt.axvline(vert_bar_at, color='r', linestyle='dashed', linewidth=1)
    mean = np.mean(arr)
    mi = np.min(arr)
    ma = np.max(arr)
    var = np.var(arr)
    n_values = len(arr)
    full_title = '%s. N: %d, Mean: %.2f, Var: %.2f, Max: %.2f, Min: %.2f' % (title, n_values, mean, var, ma, mi);
    plt.title(full_title);
    plt.show()
    
    
def show_interactive_distribution(arr, title='Data', distplot=False, bins = 20, vert_bar_at=None, figsize=[10,7]):
    fig=plt.figure()
    ax=plt.subplot()
    
    (n,bins,patches) = plt.hist(arr, bins = bins)
    if vert_bar_at:
        plt.axvline(vert_bar_at, color='r', linestyle='dashed', linewidth=1)
    mean = np.mean(arr)
    mi = np.min(arr)
    ma = np.max(arr)
    var = np.var(arr)
    n_values = len(arr)
    full_title = '%s. N: %d, Mean: %.2f, Var: %.2f, Max: %.2f, Min: %.2f' % (title, n_values, mean, var, ma, mi);
    plt.title(full_title);

    xpos=bins[:-1]

#     bars = plt.bar(xpos,arr)

    annot = ax.annotate("", xy=(0,0), xytext=(-20,20),textcoords="offset points",
                        bbox=dict(boxstyle="round", fc="black", ec="b", lw=2),
                        arrowprops=dict(arrowstyle="->"))
    annot.set_visible(False)

    def update_annot(p):
        x = p.get_x()+p.get_width()/2.
        y = p.get_y()+p.get_height()
        annot.xy = (x,y)
        text = "({:.2g},{:.2g})".format( x,y )
        annot.set_text(text)
        annot.get_bbox_patch().set_alpha(0.4)


    def hover(event):
        vis = annot.get_visible()
        if event.inaxes == ax:
            for p in patches:
                cont, ind = p.contains(event)
                if cont:
                    update_annot(p)
                    annot.set_visible(True)
                    fig.canvas.draw_idle()
                    return
        if vis:
            annot.set_visible(False)
            fig.canvas.draw_idle()

    
    fig.canvas.mpl_connect("motion_notify_event", hover)
    

    plt.show()
```

## Accuracy on targets vs time between repeats

```python
## Score vs repeat time
problematic=[]
time_dict = {}
i=0
df_nona_presentations['delay_in_vids'] = np.nan
df_nona_presentations['delay_in_ms'] = np.nan
print(len(df_nona_presentations[df_nona_presentations['vigilance']==1].groupby(['id_level','id_video'])))
for name, group in tqdm(df_nona_presentations[df_nona_presentations['targeted']==1].groupby(['id_level','id_video'])):
    i+=1
    if len(group) <2:
        print('Problem in this group!')
        display(group)
        problematic.append([group['id_level'].iloc[0],group['id_video'].iloc[0],group['position'].iloc[0]]])
        continue
        
#     print(len(group))
#     display(group)
    id_level = group['id_level'].values[0]
    id_video = group['id_video'].values[0]
    delay_in_vids = group['position'].iloc[1]-group['position'].iloc[0]
    delay_in_ms = group['start_msec'].iloc[1]-group['start_msec'].iloc[0]
    response = group['response'].iloc[1]
    time_dict[(id_video, id_level)] = (delay_in_vids, delay_in_ms, response)
```

```python
recognized_pos=[]
recognized_time=[]
non_recognized_pos=[]
non_recognized_time=[]
all_pos = [v[0] for v in time_dict.values()]
all_time = [v[1] for v in time_dict.values()]
for k,v in time_dict.items():
#     print(v)
    if v[2]:
        recognized_pos.append(v[0])
        recognized_time.append(v[1])
    else:
        non_recognized_pos.append(v[0])
        non_recognized_time.append(v[1])
    
```

```python
def compare_accuracy_across_time(all_delays, recognized_delays, n_bins=20, figsize=[10,7], title_suffix='Number of videos'):
    
    plt.figure(figsize=figsize)
    (n_all, bins, patches) = plt.hist(all_delays, bins=n_bins)
    (n_rec, bins, patches) = plt.hist(recognized_delays, bins=bins)
    plt.xlabel(title_suffix+' between first and second presentation')
    plt.ylabel('Number of videos')
    plt.legend(['all targets', 'recognized targets'])
    
    plt.figure(figsize=figsize)
    plt.bar(bins[:-1],n_rec/n_all, color='green', width=max(bins)/len(bins), align='edge')
    plt.ylabel('Accuracy on targets')
    plt.xlabel(title_suffix+' between first and second presentation')
```

```python
compare_accuracy_across_time(all_pos, recognized_pos)
```

```python
compare_accuracy_across_time(all_time, recognized_time, title_suffix='Time in ms')
```
